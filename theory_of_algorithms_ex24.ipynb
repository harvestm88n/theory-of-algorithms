```python
def find_max_edge(tree, s, f):
    if s == f:
        return -1, None

    seen = {s}
    stack = [(s, -1, None)]

    while stack:
        v, cur_max, max_e = stack.pop()

        if v == f:
            return cur_max, max_e

        for to, w in tree[v]:
            if to not in seen:
                seen.add(to)
                if w > cur_max:
                    stack.append((to, w, (v, to, w)))
                else:
                    stack.append((to, cur_max, max_e))

    return -1, None


def same(e1, e2):
    if e1 is None or e2 is None:
        return False
    a, b, w1 = e1
    x, y, w2 = e2
    return w1 == w2 and ((a == x and b == y) or (a == y and b == x))


def check(T, v, w, c):
    max_w, _ = find_max_edge(T, v, w)
    return c >= max_w


def build(edges, T, v, w, c):
    max_w, bad = find_max_edge(T, v, w)

    if c >= max_w:
        return edges

    new = []
    for e in edges:
        if not same(e, bad):
            new.append(e)
    new.append((v, w, c))
    return new


# пример
edges = [(0, 1, 2), (1, 2, 4), (2, 3, 6), (3, 4, 3)]

T = {
    0: [(1, 2)],
    1: [(0, 2), (2, 4)],
    2: [(1, 4), (3, 6)],
    3: [(2, 6), (4, 3)],
    4: [(3, 3)]}

v, w, c = 0, 3, 5

print("Пункт A:", check(T, v, w, c))
print("Пункт B:", build(edges, T, v, w, c))
```
