```python
def check_consistency(n, pairs):
    species = [-1] * n

    neighbors = []
    for i in range(n):
        neighbors.append([])

    for k in range(len(pairs)):
        a = pairs[k][0]
        b = pairs[k][1]
        relation = pairs[k][2]
        neighbors[a].append([b, relation])
        neighbors[b].append([a, relation])

    for i in range(n):
        if species[i] == -1:
            stack = [[i, 0]]

            while len(stack) > 0:
                current = stack[-1][0]
                current_species = stack[-1][1]
                stack.pop()

                if species[current] != -1:
                    if species[current] != current_species:
                        return False
                    continue

                species[current] = current_species

                for p in range(len(neighbors[current])):
                    neighbor = neighbors[current][p][0]
                    relation = neighbors[current][p][1]

                    if relation == 'same':
                        new_species = current_species
                    else:
                        new_species = 1 - current_species

                    stack.append([neighbor, new_species])

    return True


if __name__ == "__main__":
    print(check_consistency(3, [[0,1,'same'], [1,2,'different']]))
    print(check_consistency(3, [[0,1,'same'], [1,2,'same'], [0,2,'different']]))
```
