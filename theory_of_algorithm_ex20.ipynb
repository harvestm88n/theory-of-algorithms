```python
def get_best_time(people):
# people: [имя, плавание, велосипед, бег]

  people.sort(key=lambda person: person[2] + person[3],reverse=True)

  swim_total = 0
  best_time = 0

  for name, swim_time, bike_time, run_time in people:
    swim_total += swim_time
    finish_time = swim_total + bike_time + run_time
    if finish_time > best_time:
      best_time = finish_time

  return people, best_time


count = int(input("Введите количество участников: "))
people = []

for i in range(count):
  print('='*75)
  name = input("Имя участника: ")
  swim_time = int(input("Время плавания: "))
  bike_time = int(input("Время велосипеда: "))
  run_time = int(input("Время бега: "))
  people.append([name, swim_time, bike_time, run_time])

best_order, answer = get_best_time(people)

print('*'*75)
print("Оптимальный порядок участников:")

for person in best_order:
  print(person[0])

print("Минимальное время окончания:", answer)
```
