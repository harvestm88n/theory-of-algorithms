```python
def find_peak(arr):
    left = 0
    right = len(arr) - 1

    while left < right:
        mid = (left + right) // 2

        if arr[mid] < arr[mid + 1]:
            left = mid + 1
        else:
            right = mid

    return left

arr = [1, 3, 5, 7, 6, 4, 2]
peak_index = find_peak(arr)

print(f"Массив: {arr}")
print(f"Пик на индексе {peak_index}")
print(f"Значение пика: {arr[peak_index]}")
```

```python
def find_best_days(prices):
    if len(prices) < 2:
        return "Нет дней"

    min_price = prices[0]
    min_day = 0
    max_profit = 0
    buy_day = 0
    sell_day = 0

    for i in range(1, len(prices)):
        profit = prices[i] - min_price

        if profit > max_profit:
            max_profit = profit
            buy_day = min_day
            sell_day = i

        if prices[i] < min_price:
            min_price = prices[i]
            min_day = i

    if max_profit <= 0:
        return "Прибыль невозможна"
    else:
        return f"Покупать в день {buy_day + 1}, продавать в день {sell_day + 1}"


prices = [9, 1, 5]
print(find_best_days(prices))
```
